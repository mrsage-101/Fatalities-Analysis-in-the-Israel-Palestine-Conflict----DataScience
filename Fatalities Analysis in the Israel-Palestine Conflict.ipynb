## Task 1
### Exploring and Visualizing the Fatalities

```python
df['date_of_death'] = pd.to_datetime(df['date_of_death'])
df['year_of_death'] = df['date_of_death'].dt.year

df
```

#### Plot for Count of deaths against Year

```python
plt.figure(figsize=(10,6))
plt.title('Deaths Against Year')
plt.hist(x=df['year_of_death'], bins=20, ec='black', color='blue')

cnts, b_edg, _ = plt.hist(x=df['year_of_death'], bins=20, ec='black', color='blue')
for cnts, b_edg in zip(cnts, b_edg):
    plt.text(b_edg + 0.5, cnts, str(int(cnts)), ha='center', va='bottom')

plt.xlabel('Year of Death')
plt.ylabel('Number of Fatalities')
plt.ylim(0, 2500)
plt.show()
```

#### Line Plot for Spikes in Fatalities

```python
fatalities_year = df.groupby('year_of_death').size()

plt.figure(figsize=(10,6))
plt.title('Spike in Fatalities')
plt.plot(fatalities_year.index, fatalities_year, color='blue', marker='o')

for saal, cnts in zip(fatalities_year.index, fatalities_year):
    plt.text(saal, cnts, str(cnts), ha='center', va='bottom')

plt.xlabel('Year of Death')
plt.ylabel('Number of Fatalities')
plt.ylim(0,3000)
plt.grid()
plt.show()
```

#### Seaborn plot to show fatalities across 12 months in 2014

```python
data_2014 = df[df['year_of_death'] == 2014].copy()
data_2014['month'] = data_2014['date_of_death'].dt.month
monthly_death_counts = data_2014['month'].value_counts().sort_index() #no deaths in month 4
#print(monthly_death_counts)
monthly_death_counts = data_2014.groupby('month').size().reset_index(name='deaths')
all_months = pd.DataFrame({'month': range(1, 13), 'deaths': 0})
monthly_death_counts = all_months.merge(monthly_death_counts, how='left', on='month').fillna(0)
monthly_death_counts.drop(columns='deaths_x')
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=[14,6])

bars = sns.barplot(x=monthly_death_counts['month'], y=monthly_death_counts['deaths_y'], color='r')
for bar in bars.patches:
    bars.annotate(f'{bar.get_height():.0f}', (bar.get_x() + bar.get_width() / 2., bar.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.title("Deaths in 2014 across 12 months")
plt.xlabel("Months")
plt.ylabel("Fatilities")
plt.xticks(monthly_death_counts.index)
plt.ylim(0,2000)
plt.show()
```
